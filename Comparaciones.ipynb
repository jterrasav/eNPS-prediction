```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el DataFrame con los resultados
results = pd.DataFrame({
    "Modelo": [
        "MLP básico sin secuencias",
        "MLP hypertuning sin secuencias",
        "MLP básico con secuencias",
        "MLP hypertuning con secuencias",
        "LSTM básico con secuencias",
        "LSTM hypertuning con secuencias"
    ],
    "MSE": [1.6751, 1.6676, 1.5298, 1.4819, 1.4121, 1.4081],
    "MAE": [0.9192, 0.9167, 0.9024, 0.8486, 0.8239, 0.8180],
    "R2": [0.6163, 0.6181, 0.6244, 0.6362, 0.6533, 0.6543]
})
results
```

```python
# Volver a definir best_mse, best_mae, best_r2
best_mse = results["MSE"].idxmin()
best_mae = results["MAE"].idxmin()
best_r2 = results["R2"].idxmax()

# Volver a crear gráfico
metrics = ["MSE", "MAE", "R2"]
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Preparar colores para destacar mejores
for i, metric in enumerate(metrics):
    colors = ['skyblue'] * len(results)
    if metric == "MSE":
        colors[best_mse] = 'orange'
    elif metric == "MAE":
        colors[best_mae] = 'orange'
    elif metric == "R2":
        colors[best_r2] = 'orange'

    axes[i].barh(results["Modelo"], results[metric], color=colors)
    axes[i].set_title(f"{metric}")
    axes[i].invert_yaxis()
    axes[i].set_xlabel(metric)

plt.tight_layout()
plt.show()
```

```python
# Crear 3 gráficos independientes, uno por métrica

# MSE
fig, ax = plt.subplots(figsize=(6, 4))
colors = ['skyblue'] * len(results)
colors[best_mse] = 'orange'
ax.barh(results["Modelo"], results["MSE"], color=colors)
ax.set_title("MSE")
ax.invert_yaxis()
ax.set_xlabel("MSE")
plt.tight_layout()
plt.show()

# MAE
fig, ax = plt.subplots(figsize=(6, 4))
colors = ['skyblue'] * len(results)
colors[best_mae] = 'orange'
ax.barh(results["Modelo"], results["MAE"], color=colors)
ax.set_title("MAE")
ax.invert_yaxis()
ax.set_xlabel("MAE")
plt.tight_layout()
plt.show()

# R²
fig, ax = plt.subplots(figsize=(6, 4))
colors = ['skyblue'] * len(results)
colors[best_r2] = 'orange'
ax.barh(results["Modelo"], results["R2"], color=colors)
ax.set_title("R²")
ax.invert_yaxis()
ax.set_xlabel("R²")
plt.tight_layout()
plt.show()
```
